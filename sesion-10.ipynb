## Distribuciones discretas de mayor aplicación

```python
#Distribución de Bernoulli.
from scipy.stats import bernoulli

p = 0.3
data = bernoulli.rvs(p=p, size=100)
len(data)
```

```python
mean, var, skew, kurt = bernoulli.stats(p=p, moments='mvsk')

mean, var, skew, kurt
```

```python
import seaborn as sns 

ax = sns.distplot(data, bins=30, kde=False, color='blue')
ax.set(xlabel = 'Distribución de Bernoulli', ylabel='Frecuencia')
ax.axvline(x=mean, linestyle='--', label='Media', c='g')
ax.legend()
```
