# Funciones de distribución discreta y continua

## Funciones de distribución discreta

```python
#Cálculo de las probabilidades P(X = xi) para vada variable aleatoria discreta

p = 0.3
n = 8

x = list(range(0,n+1,1))
y = [] #P(xi) para todos los valores de x

for i in x:
    y.append(binom.pmf(i, p=p, n=n))

y
```

```python
fig, ax = plt.subplots()
ax.bar(x, y)
ax.set_ylabel('Probabilidad discreta')
```

```python
#Probabilidad acumulada y Regla de completitud del espacio.
#La suma de las probabilidades de las variables aleatorias es igual a 1.

np.sum(y)
```

```python
#Valor esperado condicional

media = np.average(x, weights=y)
#Promedio ponderado por el peso de cada probabilidad

varianza = np.average(((x - media)**2), weights=y)

media, varianza
```

```python
desv_std = np.sqrt(varianza)
desv_std
```

```python
fig, ax = plt.subplots()
ax.bar(x, y)
ax.set_ylabel('Probabilidad discreta')
ax.axvline(x = media, c = 'r', linestyle = '--', label = 'media')
ax.axvline(x = media + 3*desv_std, c= 'g', linestyle = '--', label = '+3 desv. std.')
ax.axvline(x = media - 3*desv_std, c= 'g', linestyle = '--', label = '-3 desv. std.')
ax.legend()
```

## Funciones de distribución continua 

```python
P = 0.3
N = 10000
x = list(range(1, N+1, 1))
y = []

for i in x:
    y.append(binom.cdf(i, p=P, n=N))
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Probabilidad continua acumulada')
```
