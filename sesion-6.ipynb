# Visualización de datos

```python
import scipy.stats
import numpy as np
import pandas as pd 

import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('bicicletas-compartidas.csv')
df.head()
```

## Variables categóricas

```python
df.columns
```

```python
y = df['cuartil-ano'].apply(lambda x: 'cuartil-' + str(int(x)))
y.head()
```

```python
# Diagrama de barras

fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuartiles del año')
ax.set_ylabel('Frecuencia')
```

```python
fig, ax = plt.subplots()
ax.bar(y.value_counts().index, y.value_counts())
ax.set_xlabel('Cuartiles del año')
ax.set_ylabel('Frecuencia')
ax.patches[2].set_facecolor('red')
```

```python
# Diagrama de pie

fig, ax = plt.subplots()
ax.pie(y.value_counts(), labels = y.value_counts().index, autopct='%1.1f%%')
ax.set_title('Diagrama de pie')
```

## Variables numéricas

```python
y = df['viento']
```

```python
#Histograma

fig, ax = plt.subplots()
ax.hist(y, bins = 30)
ax.set_xlabel('Viento')
ax.set_ylabel('Frecuencias')
plt.axvline(np.mean(y), c = 'r', linestyle = '--', label = 'Media') #Graficando la media
plt.axvline(np.mean(y)+np.std(y), c = 'g', linestyle = '--', label='+desv. est.') #Graficando lado izquierdo desv. std
plt.axvline(np.mean(y)-np.std(y), c = 'g', linestyle = '--', label='-desv. est.')
ax.legend() #Imprimir legendas de lineas
ax.set_title('Histograma de distribución del viento') #Añdirle título a la gráfica
ax.grid(color='gray', linestyle = '--', linewidth='0.5')#Añadir un grid
```

```python
# Boxplot

y = df['bicis-compartidas']

fig, ax = plt.subplots()
ax.boxplot(x = y)
```

```python
fig, ax = plt.subplots()
sns.boxplot(x = 'cuartil-ano', y = 'bicis-compartidas', data = df)
```

```python
# Scatterplot o Gráfica de dispersión

fig, ax = plt.subplots()
ax.scatter(df['viento'], df['bicis-compartidas'], alpha = 0.03)
ax.set_title('Distribución conjunta de viento y bicis-compartidas')
ax.set_xlabel('Viento')
ax.set_ylabel('Bicis-compartidas')
```
